## <b>1. Introduction</b>

<div style="text-align: justify;">
<b>Malaria is a severe and sometimes fatal disease</b> caused by Plasmodium parasites, which are transmitted to humans through the bites of infected female Anopheles mosquitoes. The parasites invade and damage red blood cells, causing symptoms that can range from mild to life-threatening. Mild symptoms includes chills and headache, while severe symptoms include fatigue, confusion, seizures, and difficulty breathing. This disease poses a significant health risk, especially in regions with high transmission rates, as it can remain asymptomatic for over a year. Globally in 2022, there were an estimated 249 million malaria cases and 608,000 malaria deaths in 85 countries (WHO, 2022). The African Region carries a disproportionately high share of the global malaria burden. Children under 5 accounted for about 78% of all malaria deaths in the region (WHO, 2022). <b>Traditional malaria diagnosis relies on the expertise of professionals to inspect blood samples under a microscope</b>, which is time-consuming and subject to human error and variability. A vaccine to prevent malaria is available; however, its variable efficacy underscores the need for new interventions that offer high-level protection against disease. Scientists are researching improved vaccines and preventive interventions as well as mosquito control techniques, easy-to-use diagnostics, and improved therapies as parasites continue to develop resistance to currently available antimalarials (NIAID, 2019).
</div><br>

<div style="text-align: justify;">
More recently, malaria Rapid Diagnostic Tests (RDTs) were developed to assist in the diagnosis of malaria by providing evidence of the presence of malaria parasites in human blood. RDTs are an alternative to diagnosis based on clinical grounds or microscopy, particularly where good quality microscopy services cannot be readily provided (WHO, 2023). However, based on the WHO evaluations, RDTs for malaria show varying performance metrics. On average, RDTs demonstrate a sensitivity (recall) of around 75-90% and a specificity (precision) of approximately 90-95% (WHO, 2024). This means that while RDTs are generally effective in identifying malaria infections, there is a risk of false negatives, particularly in cases with low parasite density. These metrics highlight the importance of using RDTs in conjunction with other diagnostic methods to ensure accurate malaria diagnosis and management.
</div><br>

<div style="text-align: justify;">
Given these challenges, an automated system for early and accurate detection of malaria is crucial. Such a system can significantly improve diagnostic accuracy, reduce the time required for diagnosis, and minimize human error. ### <b>3.2. Loading the Data</b>

```python
# Defining base directory dynamically to avoid hardcoding absolute paths
base_dir = os.path.join(os.getcwd(), 'datasets', 'malaria_detection')

# Creating directories if they do not exist
os.makedirs(base_dir, exist_ok=True)

# Path to the ZIP file
zip_path = os.path.join(base_dir, 'cell_images.zip')

# Extracting the ZIP file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(base_dir)
    
print(f"Data extracted to: {base_dir}")
```

<b>Observations and Insights:
- The extracted folder has different folders for train and test data will contain the different sizes of images for parasitized and uninfected cells within the respective folder name.
- The provided images must be loaded, resized to an uniform size of 64x64 pixels, and converted to 4D arrays so that they can be used as an input for the CNN.
- Also, we need to create the labels for both types of images to be able to train and test the model.

```python
from PIL import Image

# Function to validate image files
def validate_images(directory):
    """
    Check for corrupted or unreadable image files in a directory.
    
    Args:
        directory (str): Path to the directory containing images.
        
    Returns:
        list: A list of invalid image file paths.
    """
    invalid_images = []
    for img_file in os.listdir(directory):
        img_path = os.path.join(directory, img_file)
        try:
            img = Image.open(img_path)
            img.verify()  # Verify image integrity
        except Exception as e:
            invalid_images.append(img_path)
    return invalid_images

# Directories for train and test data
train_dir = os.path.join(base_dir, 'cell_images', 'train')
test_dir = os.path.join(base_dir, 'cell_images', 'test')

# Validate images in train and test directories
parasitized_dir_train = os.path.join(train_dir, 'parasitized')
uninfected_dir_train = os.path.join(train_dir, 'uninfected')
parasitized_dir_test = os.path.join(test_dir, 'parasitized')
uninfected_dir_test = os.path.join(test_dir, 'uninfected')

invalid_train = validate_images(parasitized_dir_train) + validate_images(uninfected_dir_train)
invalid_test = validate_images(parasitized_dir_test) + validate_images(uninfected_dir_test)

print(f"Corrupted images in training set: {len(invalid_train)}")
print(f"Corrupted images in test set: {len(invalid_test)}")

if invalid_train or invalid_test:
    print("Corrupted image paths:")
    print(invalid_train + invalid_test)
```

```python
# Loading, resizing and converting training images to a numpy array 

# Directories for parasitized and uninfected images in train data
parasitized_dir_train = os.path.join(train_dir, 'parasitized')
uninfected_dir_train = os.path.join(train_dir, 'uninfected')

# Initializing lists to hold images and labels
train_images = []
train_labels = []

# Function to process images in a directory
def process_images_from_directory(directory, label, images, labels):
    """
    Load, resize, and convert images to numpy arrays.
    
    Args:
        directory (str): Path to the image directory.
        label (int): Label corresponding to the image class.
        images (list): List to store image arrays.
        labels (list): List to store labels.
    """
    for filename in os.listdir(directory):  # Iterate over all files in the specified directory
        if filename.endswith('.png'):       # Check if the file is a PNG image
            img_path = os.path.join(directory, filename)  # Construct the full path to the image file
            img = load_img(img_path, target_size=(64, 64))  # Load and resize the image
            img_array = img_to_array(img)  # Convert image to a numpy array
            images.append(img_array)       # Append the image array to the image list
            labels.append(label)           # Append the corresponding label to the label list

# Loading parasitized images and label = 1
process_images_from_directory(parasitized_dir_train, 1, train_images, train_labels)

# Loading uninfected images and label = 0
process_images_from_directory(uninfected_dir_train, 0, train_images, train_labels)

# Converting lists to numpy arrays
train_images = np.array(train_images)
train_labels = np.array(train_labels)
```

```python
# Reproducing the same code for the test data

# Directories for parasitized and uninfected images in test data
parasitized_dir_test = os.path.join(test_dir, 'parasitized')
uninfected_dir_test = os.path.join(test_dir, 'uninfected')

# Initializing lists to hold images and labels
test_images = []
test_labels = []

# Loading parasitized images and label = 1
process_images_from_directory(parasitized_dir_test, 1, test_images, test_labels)

# Loading uninfected images and label = 0
process_images_from_directory(uninfected_dir_test, 0, test_images, test_labels)

# Converting lists to numpy arrays
test_images = np.array(test_images)
test_labels = np.array(test_labels)
```

### <b>3.5. Processing Images using Gaussian Blurring</b>

#### <b> Gaussian Blurring on Train Data</b>

```python
# Initializing an empty list to hold the blurred images for train data
train_images_blurred = []

# Looping through each image in the train data
for img in train_images_normalized:
    img_blurred = cv2.GaussianBlur(img, (5, 5), 0)      # Applying Gaussian Blurring with a kernel size of 5x5
    train_images_blurred.append(img_blurred)            # Appending the blurred image to the list

# Converting the list to a numpy array
train_images_blurred = np.array(train_images_blurred)
```

### <b>4.3. Base Model</b>

#### <b>4.3.1. Building the Model</b>

```python
# Clearing the backend and fixing the seed to ensure the same output every time
backend.clear_session()

np.random.seed(42)
random.seed(42)
tf.random.set_seed(42)
```

```python
# Initializing the sequential model
model = Sequential()

# Adding the first convolutional layer with 64 filters, kernel size of 3x3,
# tanh activation, and input shape of 64x64x3 followed by a MAxPooling2D
# layer to reduce the size of output of first Conv2D layer
model.add(Conv2D(64, (3, 3), padding='same', activation='tanh', input_shape=(64, 64, 3)))
model.add(MaxPooling2D((2, 2), padding = 'same'))

# Adding the second convolutional layer with 32 filters and kernel size of 3x3, tanh activation
# followed by a MaxPooling2D layer to reduce the size of output of first Conv2D layer
model.add(Conv2D(32, (3, 3), padding='same', activation='tanh'))
model.add(MaxPooling2D((2, 2), padding = 'same'))

# Adding the third convolutional layer with 16 filters and kernel size of 3x3, tanh activation
# followed by a MaxPooling2D layer to reduce the size of output of first Conv2D layer
model.add(Conv2D(16, (3, 3), padding='same', activation='tanh'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Flatten the 3D output to 1D
model.add(Flatten())

# Adding a dense (fully connected) layer with 64 nodes and tanh activation
model.add(Dense(64, activation='tanh'))

# Adding a dropout layer with a dropout rate of 0.25 for regularization
model.add(Dropout(0.25))

# Adding the output layer with 2 units (for binary classification) and sigmoid activation
model.add(Dense(2, activation='sigmoid'))

# Printing the summary of the CNN model
model.summary()
```

<b>Comments and Insights:</b>
- The architecture of the first proposed model (<i>Model</i>) consists of 6 layers in the feature-extraction group (3 pairs of Conv2D + MaxPooling2D layers) followed by 3 layers in the classification group.
- In between, a Flatten layer was added to convert the 3D matrix of features produced by the feature-exxtraction group into a 1D vector required as input for the classification group of layers.
- The resulting model has a total of 90,610 trainable parameters

<b>Compiling the Model</b>

```python
# Compiling the model with binary crossentropy loss, Adam optimizer, and accuracy as the metric
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

### <b>4.5. Pre-trained model (VGG16)</b>

<div style="text-align: justify;">
For the next model we will apply the idea of Transfer Learning, which reuses a pre-trained model on a new problem. This approach allows training deep neural networks with comparatively little data. This is very useful in the data science field since most real-world problems typically do not have millions of labeled data points to train such complex models.
</div><br>

<div style="text-align: justify;">
Here, we will use a pre-built architecture with its pre-trained weights and biases from VGG16, which was trained on the ImageNet dataset.
</div>

```python
# Importing necessary libraries
from tensorflow.keras import Model
from tensorflow.keras.applications.vgg16 import VGG16
```

```python
# Clearing the backend and fixing the seed to ensure the same output every time
backend.clear_session()

np.random.seed(42)
random.seed(42)
tf.random.set_seed(42)
```

#### <b>4.5.1. Importing the VGG16 Model</b>

```python
# Importing the VGG16 model
vgg_model = VGG16(weights = 'imagenet',      # Using pre-trained weights from the ImageNet dataset
                  include_top = False,       # Importing the model without the fully connected layers 
                  input_shape =(64, 64 ,3),  # Defining the input shape
                  pooling='max'              # Reducing the output of the convolutional layers to a 1D vector
)

# Checking the summary of the model
vgg_model.summary()
```

#### <b>4.5.2. Building the Model</b>

```python
# Taking the output of the 5th block of the VGG16 model as the input of our new model
transfer_layer = vgg_model.get_layer('block5_pool')
```

```python
# We will not train the layers from the VGG16 model. We will use the pre-trained weights and biases,
# so we will make all the layers of the VGG model non-trainable
vgg_model.trainable = False
```

```python
# Adding custom classification layers on top of the VGG16 base model using Functional Model API

# Flattenning the output from the 'block5_pool' layer
x = Flatten()(transfer_layer.output)       

# # Adding the first dense layer with 256 nodes and ReLU activation
x = Dense(256, activation='relu')(x)

# Adding the second dense layer with 128 nodes and ReLU activation
x = Dense(128, activation='relu')(x)

# Adding a dropout layer with a dropout rate of 0.5 for regularization
x = Dropout(0.5)(x)

# Adding the third dense layer with 64 nodes and ReLU activation
x = Dense(64, activation='relu')(x)

# Adding a BatchNormalization layer
x = BatchNormalization()(x)

# Adding the output layer with 2 units and softmax activation
pred = Dense(2, activation='softmax')(x)

# Initializing the new model
model_4 = Model(inputs = vgg_model.input, outputs = pred)

# Printing the summary of the model
model_4.summary()
```

<b>Comments and Insights:</b>
- The base architecture of <i>Model 4</i> was imported from VGG16's feature extraction layers, including their pre-trained weights and biases. This architecture consists of 5 blocks, each containing 2 Conv2D layers followed by a MaxPooling2D layer.
- On top of this pre-built architecture, we added a classification group consisting of 6 layers.
- In between, a Flatten layer was added to convert the 3D matrix of features produced by the feature-exxtraction group into a 1D vecto r required as input for the classification group of layers.
- The resulting model comprises a total of 15,280,770 parameters, with 14,714,816 (96.3%) being non-trainable (inherited from VGG16) and 565,954 (3.7%) being trainable (from the classification layers).

<b>Compiling the Model</b>

```python
# Compiling the model with binary crossentropy loss, Adam optimizer, and accuracy as the metric
opt = tf.keras.optimizers.Adam(learning_rate=0.001)
model_4.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
```

<b>Using Callbacks</b>

```python
# Defining callbacks to stop the training if the accuracy does not improve after certain number of epochs
# and to save the best model after every iteration
stop = EarlyStopping(monitor = 'val_loss', mode = 'min', verbose = 1, patience = 5)

checkpoint = ModelCheckpoint('best_model_4.keras', monitor = 'val_accuracy', 
                             mode = 'max', verbose = 1, save_best_only = True)
```

<b>Fitting and Training the Model</b>

We will use the training data after Data Augmentation because when fine-tuning a pre-trained model like VGG16, the lower layers typically extract general features (e.g., edges, textures), while the upper layers adapt to more specific features of the new task. Data augmentation ensures that the feature extraction process remains effective across a wider range of data variations, facilitating better transfer learning.

Mixup is an advanced data augmentation technique that generates new training examples by combining pairs of images and their labels, thus introducing greater variability and richness into the training data. This could help the model learning more generalized features and reducing the risk of overfitting. By blending images and labels, Mixup could strengthen the model's robustness and improve its accuracy and reliability on unseen data, making it an ideal enhancement for further performance of <i>Model 3</i> in malaria detection. 2. <b>Ensemble Learning with Data Augmentation and Mixup.</b> Two separate models can be trained independently using different techniques (<i>Model 3</i> with Data Augmentation and <i>Model 6</i> with Mixup). The final predictions are combined using averaging as ensemble method, leveraging the strengths of both models. ### <b>5.1. Fine-Tunning Model 3 using Mixup</b>

<div style="text-align: justify;">
Mixup augmentation technique generates new training samples by linearly interpolating between two images and their associated labels. The blending of images could act as a regularization technique, helping to prevent overfitting by forcing the model to learn more generalized features rather than memorizing specific examples. In addition, by training on mixed images, the model potentially becomes more robust to variations and potential noise in the real-world data, thereby improving its performance on unseen data, and can lead to a better generalization, as the model is exposed to a wider rnge of possible image variations during training.
</div>

#### <b>5.1.1. Applying Mixup Augmentation to a Batch of Images and Labels</b>

```python
# Defining the mixup function to apply Mixup augmentation to a batch of images and labels
def mixup(images, labels, alpha=0.2):
    batch_size = images.shape[0]                     # Getting the batch size from the number of images
    lam = np.random.beta(alpha, alpha, batch_size)   # Sampling lambda values from a Beta distribution
    lam = np.maximum(lam, 1 - lam)                   # Ensuring lambda is not greater than 1 - lambda
    indices = np.random.permutation(batch_size)      # Generating a random permutation of the batch indices

    # Creating the mixup images using the sampled lambda values
    mixed_images = lam[:, None, None, None] * images + (1 - lam)[:, None, None, None] * images[indices]
    # Creating the mixup labels using the sampled lambda values
    mixed_labels = lam[:, None] * labels + (1 - lam)[:, None] * labels[indices]
    
    return mixed_images, mixed_labels

#  Applying Mixup augmentation to the training images and labels
mixed_images, mixed_labels = mixup(train_images_normalized, train_labels_encoded)
```

```python
# Defining the Mixup generator function
def mixup_generator(mixed_images, mixed_labels, batch_size):
    num_samples = mixed_images.shape[0]
    while True:
        indices = np.random.choice(num_samples, batch_size, replace=False)
        yield mixed_images[indices], mixed_labels[indices]

# Creating the Mixup generator with batch size as 64
batch_size = 64
mixup_gen = mixup_generator(mixed_images, mixed_labels, batch_size)
```

```python
# Plotting a random example of images after Mixup
def plot_mixup_images(mixed_images, mixed_labels, num_images=8):
    fig, axes = plt.subplots(2, 4, figsize=(10, 6))
    for ax in axes.flatten():
        idx = np.random.randint(0, mixed_images.shape[0])
        image = mixed_images[idx]
        label = mixed_labels[idx]
        ax.imshow(image)
        ax.set_title(f'Mixed Label:\n[{label[0]:.2f}, {label[1]:.2f}]')
        ax.axis('off')
    fig.tight_layout()
    plt.show()

plot_mixup_images(mixed_images, mixed_labels)
```

<b>Comments and Insights:</b>
- The resulting images from Mixup appear as blended versions of two original images. This blending effect introduces variability in the training dataset, which is essential for improving the generalization capabilities of the model.
- The labels are also mixed, as shown by the interpolated label values. For instance, an image with a mixed label of (0.02, 0.98) indicates that it is primarily a parasitized image with a slight influence from an uninfected image.

- The resulting model has a total of 50,706 trainable parameters

```python
# Compiling the model with binary crossentropy loss, Adam optimizer, and accuracy as the metric
opt = tf.keras.optimizers.Adam(learning_rate=0.001)
model_6.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
```

<b>Using Callbacks</b>

```python
# Defining callbacks to stop the training if the accuracy does not improve after certain number of epochs
# and to save the best model after every iteration
stop = EarlyStopping(monitor = 'val_loss', mode = 'min', verbose = 1, patience = 5)

checkpoint = ModelCheckpoint('best_model_6.keras', monitor = 'val_accuracy', 
                             mode = 'max', verbose = 1, save_best_only = True)
```

```python
# Fitting the model with 30 epochs and batch size as 32
history_model_6 = model_6.fit(mixed_images, mixed_labels,
                              epochs = 30,
                              batch_size = 32,
                              validation_split = 0.2,
                              callbacks = [stop, checkpoint],
                              verbose = 1)
```

The emergence of <b>Deep Learning techniques provides an opportunity to automate and improve the accuracy of malaria detection</b> from blood cell images, reducing the burden on healthcare systems and potentially saving lives.

To address this challenge, a variety of deep learning models and methods were explored to classify red blood cell images as either parasitized (infected with malaria) or uninfected. Simple CNN architectures, models incorporating different layers such as dense and pooling layers, and more complex networks with advanced regularization techniques were investigated. When designing and evaluating the effectiveness of the CNN models for malaria detection, several key characteristics and limitations were carefully considered at the outset of the project. One primary consideration was the quality and diversity of the dataset, which consisted of microscopic images of red blood cells. Ensuring that the dataset was representative of various conditions and stages of malaria infection was crucial for training a robust model. Additionally, the images needed to be preprocessed to a consistent size and format, which involved resizing and normalization to improve model performance and reduce computational complexity. 
</div><br>

<div style="text-align: justify;">
The architecture of the CNNs were also thoughtfully designed, balancing the depth and complexity of the model to capture intricate features while avoiding overfitting. It was essential to evaluate the model’s performance using metrics such as accuracy, precision, recall, and F1-score, ensuring that the model could reliably distinguish between parasitized and uninfected cells. In the case of malaria detection, precision and recall are critical metrics due to the severe implications of both false positives and false negatives. Balancing these metrics is essential to ensure the model’s effectiveness and reliability in clinical settings, where accurate and timely diagnosis is critical for patient care and for managing and containing drug-resistant malaria.
</div><br> 

<div style="text-align: justify;">
High precision ensures that the model accurately identifies parasitized cells, minimizing the number of false positives and thus preventing unnecessary treatments. This is especially important in countries with lower economic resources, such as many African nations, where the economic cost of unnecessary treatments can significantly strain already limited healthcare budgets. However, in this context, high recall is of paramount importance because it ensures that the model correctly identifies the majority of parasitized cells, thereby reducing the risk of false negatives. This reduction is crucial as false negatives can lead to untreated infections, which may cause severe health complications and contribute to the spread of the disease. This is especially critical for vulnerable populations, such as children under 5 years of age, where a false negative diagnosis can be fatal due to their heightened susceptibility to severe malaria complications.
</div><br>

The final proposed solution is an <b><i>Ensemble Model</i></b> combining two Convolutional Neural Networks (CNN) optimized through two different techniques of Data Augmentation. Data Augmentation in <i>Model 3</i> was performed through random transformations, whereas Data Augmentation in <i>Model 6</i> was performed using the Mixup technique. This ensemble approach addresses the specific challenges of malaria detection from microscopic images by <b>leveraging the strengths of both models</b>. The architecture of both models includes four convolutional layers, each followed by batch normalization and max pooling layers. These convolutional layers are critical for extracting intricate features from the images, enabling the model to distinguish between parasitized and uninfected cells with high accuracy. Batch normalization is applied after each convolutional layer to stabilize and accelerate the training process, ensuring the model converges more quickly and reliably. Max pooling layers reduce the spatial dimensions of the feature maps, which not only decreases computational complexity but also helps in capturing the most salient features. Additionally, both models incorporates fully connected (dense) layers with ReLU activation functions to learn complex patterns and relationships in the extracted features. Dropout layers are included in the fully connected sections to prevent overfitting by randomly omitting 50% of the input units during training. This regularization technique is crucial for enhancing the model’s ability to generalize to new, unseen data. The final layer is a softmax output layer with two units, which provides probabilistic predictions for the binary classification task of identifying parasitized versus uninfected cells. This carefully designed architecture, along with extensive hyperparameter tuning, ensures that the ensemble model achieves an <b>accuracy of 99%</b>.  The model also maintained high performance with an accuracy of 98.6% when tested with suboptimal quality images to replicate real-world conditions, indicating <b>robustness to variations in image quality</b>.
